## Load the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.discriminant_analysis import StandardScaler




a1_synthetic_path = '../Data/A1-synthetic.txt'
a1_synthetic_data = pd.read_csv(a1_synthetic_path, sep="\t") 

# display the first few rows of the dataset
print("First few rows of the dataset:")
display(a1_synthetic_data.head())

# display dataset shape and column names
print(f"Dataset shape: {a1_synthetic_data.shape}")
print(f"Column names: {a1_synthetic_data.columns.tolist()}")

```

## Normalize the data

```python
# initializing the StandardScaler for the synthetic data
scaler_synthetic = StandardScaler()

# fitting the scaler to the synthetic data and transforming it
normalized_synthetic_data = scaler_synthetic.fit_transform(a1_synthetic_data)

# converting the normalized data back to a DataFrame for easier interpretation
normalized_synthetic_df = pd.DataFrame(normalized_synthetic_data, columns=a1_synthetic_data.columns)

# displaying the first few rows of the normalized synthetic data
normalized_synthetic_df.head()
```

## Save the normalized data

```python
# save the normalized dataset to a new CSV file
normalized_synthetic_df.to_csv('A1-synthetic-normalized.csv', index=False, sep='\t')
print("Normalized dataset saved as 'A1-synthetic-normalized.csv'.")
```
